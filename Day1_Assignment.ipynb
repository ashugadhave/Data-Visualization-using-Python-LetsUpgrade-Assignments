## Line Chart

```python
x=np.arange(0,10)
y=np.arange(11,21)
x
```

```python
plt.plot(x,y,'go--',linewidth=2,markersize=12)
plt.title('Line Plot')
plt.xlabel('x-Axis')
plt.ylabel('Y-Axis')
```

```python
x=np.arange(1,11)
y=3 * x + 5
plt.title("Line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y)
```

```python
x=np.arange(0,4 * np.pi,0.1)
y=np.sin(x)
plt.title("sine waveframe")
plt.plot(x,y,color='b')
plt.show()
```

```python
days = [1,2,3,4,5,6,7]
sales_1 = [160, 150, 140, 145, 175, 165, 180]
sales_2 = [70, 90, 160, 150, 140, 145, 175]
plt.figure(figsize=(6,5),dpi=150)
plt.plot(days,sales_1)
plt.plot(days,sales_2)

plt.xlabel("Days of week")
plt.ylabel("Sales")
plt.grid()
plt.show()
```
